## Data Loading & Cleaning

```python
#loading the mvp data frame for cleaning

mvps= pd.read_csv('mvp.csv')
mvps.head()
```

Choosing the appropriate column for the MVPs dataframe, which involves feature selection.

```python
mvps=mvps[['Player', 'year', 'Pts Won', 'Pts Max', 'Share']]
mvps
```

```python
# loading the player dataframe to clean

players=pd.read_csv('playe.csv')
players.head()
```

Removing unnecessary columns for machine learning

```python
del players['Unnamed: 0']
```

```python
del players['Rk']
```

```python
players.head()
```

Identify the names to ensure they are in a proper structure with no extra special symbols.

```python
players['Player'].head(50)
```

In the list of names above, names like 'Larry Bird*' and 'Charles Barkley*' contain special symbols '*' that need to be cleaned.

```python
players['Player']=players['Player'].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row=df[df['Tm']=='TOT']
        row['Tm']=df.iloc[-1, :]['Tm']
        return row
    
players = players.groupby(['Player', 'year']).apply(single_row)
```

```python
players.head(20)
```

```python
players.index=players.index.droplevel()
```

```python
players.index=players.index.droplevel()
```

```python
players.head()
```

Combining the two dataframes into a single dataframe to facilitate future machine learning purposes.

```python
combined = players.merge(mvps, how="outer", on=['Player', 'year'])
```

```python
combined.head(10)
```

```python
combined[combined['Pts Won'] > 0]
```

To identify if there are any null values and replace them with 0

```python
combined[['Pts Won', 'Pts Max', 'Share']] = combined[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

```python
combined[combined['Pts Won'] > 0]
```

```python
combined.head()
```

```python
# loading the teams dataframe

team=pd.read_csv('teams_d.csv')

team.head(20)
```

```python
team=team[~team['W'].str.contains('Division')]
team.head(20)
```

```python
team['Team']=team['Team'].str.replace('*','', regex=False)
team.head()
```

Loading the 'nickname.txt' file for team name correction, which is particularly helpful when joining the dataframes with each other.

```python
nick = pd.read_csv('nicknames.txt', sep=',')
nick.to_csv('nickname.csv')
```

```python
nicknames={}

with open('nickname.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines[1:]:
        values = line.replace("\n", "").split(",")
        if len(values) >= 2:
            abr, name = values[1], values[2]
            nicknames[abr] = name
        else:
            # Handle lines with fewer values or empty lines
            print(f"Skipping line: {line}")
```

```python
combined['Team']=combined['Tm'].map(nicknames)
combined.head()
```

```python
stats=combined.merge(team, how='outer', on=['Team','year'])
stats.shape
```

```python
combined.shape
```

```python
del stats['Unnamed: 0']
```

```python
stats.dtypes
```

```python
stats=stats.apply(pd.to_numeric, errors='ignore')
stats.dtypes
```

```python
stats['GB'].unique()
```

```python
stats['GB']=stats['GB'].str.replace('â€”','0')
stats['GB'].unique()
```

```python
stats['GB']=pd.to_numeric(stats['GB'])
stats.dtypes
```

```python
stats.to_csv('player_mvp_stats.csv')
```
